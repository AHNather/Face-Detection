```python
import pathlib
import cv2
import matplotlib.pyplot as plt
```

```python
# Set the path to the Haar cascade
cascade_path = "/Users/ahmednather/Downloads/haarcascade_frontalface_default.xml"
cls = cv2.CascadeClassifier(cascade_path)

# Load the image
image = cv2.imread("/Users/ahmednather/Downloads/maskw.jpeg")

# Ensure the image is loaded correctly
if image is None:
    raise IOError("Failed to load image. Check the file path!")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = cls.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags=cv2.CASCADE_SCALE_IMAGE
)

# Draw rectangles around detected faces
for (x, y, width, height) in faces:
    cv2.rectangle(image, (x, y), (x + width, y + height), (255, 0, 0), 2)

# Convert BGR to RGB for display
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(rgb_image)
plt.axis('off')  # Hide axes for a cleaner image
plt.title("Detected Faces")
plt.show()
```
